```python
countries_df = pd.read_csv("dictionary.csv")
countries_df.head()
```

```python
summer_df = pd.read_csv("summer.csv")
summer_df['Season'] = 'Summer'
winter_df = pd.read_csv("winter.csv")
winter_df['Season'] = 'Winter'
summer_df = summer_df.sort_values(by="Country", ascending=True)
summer_df.head()
winter_df.head()
```

```python
summer_df = summer_df.rename(index=str, columns={"Country": "Code"})
winter_df = winter_df.rename(index=str, columns={"Country": "Code"})
```

```python
summer_countries_df = countries_df.merge(summer_df, on='Code')
summer_countries_df.head()
```

```python
winter_countries_df = countries_df.merge(winter_df, on='Code')
winter_countries_df.head()
```

```python
all_df = pd.concat([summer_countries_df, winter_countries_df], axis="index", sort=False)
all_df.tail()
```

```python
Year = all_df['Year'] >= 1984
all_df[Year]
```

```python
meda_df = all_df[Year].groupby(["Country"]).count()
```
